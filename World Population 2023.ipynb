```python
import pandas as pd 
```

```python
df = df.replace(',','', regex=True)
```

```python
df['area'] = df['area'].str.replace('M','').str.replace('K','')
df
```

```python
df['land_area'] = df['land_area'].str.replace('M','').str.replace('K','')
df
```

```python
df = df.replace('%','', regex=True)
```

```python
df = df.replace('< 1','', regex=True)
```

```python
df.tail()
```

```python
df['current_population'] = pd.to_numeric(df['current_population'])
df['population_2022'] = pd.to_numeric(df['population_2022'])
df['area'] = pd.to_numeric(df['area'])
df['land_area'] = pd.to_numeric(df['land_area'])
df['density'] = pd.to_numeric(df['density'])
df['growth_rate'] = pd.to_numeric(df['growth_rate'])
df['world_percentage'] = pd.to_numeric(df['world_percentage'])
```
